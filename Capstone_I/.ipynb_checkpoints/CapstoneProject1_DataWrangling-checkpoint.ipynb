```python
#Creating maps to normalize answers
work_interfere_1_map = {"Not applicable to me":"na"}
work_interfere_2_map = {'Never':'No', 'Rarely': 'No', 'Sometimes': 'Yes', 'Often':'Yes'}
care_options_map = {'I am not sure': 'Not sure'}
wellness_program_map = {'I don\'t know': 'Don\'t know'}
seek_help_map = {'I don\'t know': 'Don\'t know'}
anonymity_map = {'I don\'t know': 'Don\'t know'}
coworkers_map = {'Some of them': 'Maybe'}
supervisor_map = {'Some of them': 'Maybe'}
mental_vs_physical_map = {'I don\'t know': 'Don\'t know'}
obs_consequence_map = {'Yes, I observed': 'Yes', 'Yes, I experienced':'Yes'}
treatment_map = {0: 'No', 1: 'Yes'}
tech_company_map = {0: 'No', 1: 'Yes'}
self_employed_map = {0: 'No', 1: 'Yes'}
Country_map = {'United States of America': 'United States'}

#using maps to normalize answers
for i in dfs:
    i['care_options'] = i['care_options'].map(care_options_map).fillna(i['care_options'])
    i['wellness_program'] = i['wellness_program'].map(wellness_program_map).fillna(i['wellness_program'])
    i['seek_help'] = i['seek_help'].map(seek_help_map).fillna(i['seek_help'])
    i['anonymity'] = i['anonymity'].map(anonymity_map).fillna(i['anonymity'])
    i['coworkers'] = i['coworkers'].map(coworkers_map).fillna(i['coworkers'])
    i['supervisor'] = i['supervisor'].map(supervisor_map).fillna(i['supervisor'])
    i['mental_vs_physical'] = i['mental_vs_physical'].map(mental_vs_physical_map).fillna(i['mental_vs_physical'])
    i['obs_consequence'] = i['obs_consequence'].map(obs_consequence_map).fillna(i['obs_consequence'])
    i['treatment'] = i['treatment'].map(treatment_map).fillna(i['treatment'])
    i['tech_company'] = i['tech_company'].map(tech_company_map).fillna(i['tech_company'])
    i['Country'] = i['Country'].map(Country_map).fillna(i['Country'])
    
dfs[0]['work_interfere'] = dfs[0]['work_interfere'].map(work_interfere_1_map).fillna(i['work_interfere'])
dfs[1]['work_interfere'] = dfs[1]['work_interfere'].map(work_interfere_1_map).fillna(i['work_interfere'])
dfs[2]['work_interfere'] = dfs[2]['work_interfere'].map(work_interfere_2_map).fillna(i['work_interfere'])
dfs[3]['work_interfere'] = dfs[3]['work_interfere'].map(work_interfere_2_map).fillna(i['work_interfere'])

dfs[3]['self_employed'] = dfs[3]['self_employed'].map(self_employed_map).fillna(i['self_employed'])
```

```python
#normalizing gender answers
for i in dfs:   
    for index, row in i.iterrows():
        g = i.loc[index, 'Gender']
        #first, check if string, if so set to lowercase
        if isinstance(g, str):
            g = g.lower()
            #check first letter to set value
            if g[0] == 'm':
                i.loc[index, 'Gender'] = 'male'
            elif g[0] == 'f':
                i.loc[index, 'Gender'] = 'female'
            #if it doesn't being with 'm' or 'f' set to 'other'
            else:
                i.loc[index, 'Gender'] = 'other'
        #if nan, set to 'other'
        else:
            i.loc[index, 'Gender'] = float('nan')
```

```python
#all data normalized - time to merge
df = dfs[0].merge(dfs[1], 'outer')
df = df.merge(dfs[2], 'outer')
df = df.merge(dfs[3], 'outer')
##verifying combined data
print(df.info())
```

```python
#setting nan ages to avg
#iterating through rows
for index, row in df.iterrows():
    x = df.loc[index]
    if np.isnan(x['Age']):
        df.loc[index, 'Age'] = df['Age'].mean()
```

```python
#getting dummy vars
df_dummies = pd.get_dummies(df, dummy_na=True)
```
