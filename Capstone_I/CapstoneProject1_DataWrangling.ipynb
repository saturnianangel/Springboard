```python
#Gender Answers varied, dict to map not as viable. Using for-loop to map numeric values
for i in dfs:   
    for index, row in i.iterrows():
        g = i.loc[index, 'Gender']
        #first, check if string, if so set to lowercase
        if isinstance(g, str):
            g = g.lower()
            #check first letter to set numeric value
            if g[0] == 'm':
                i.loc[index, 'Gender'] = 0
            elif g[0] == 'f':
                i.loc[index, 'Gender'] = 1
            #if it doesn't being with 'm' or 'f' set to 'other'
            else:
                i.loc[index, 'Gender'] = 2
        #if nan, set to 'other'
        else:
            i.loc[index, 'Gender'] = 2
```
